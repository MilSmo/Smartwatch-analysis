```python
#avaerage steps 
avg_steps = df['TotalSteps'].mean()
```

```python
#active minutes graphs
activity_df = df[['LightlyActiveMinutes','FairlyActiveMinutes','VeryActiveMinutes']].copy()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5)) # set figure size and create subplots

for i, feature in enumerate(activity_df.columns):
    sns.set_palette("pastel")
    sns.histplot(data=activity_df, x=feature, kde=True, ax=axs[i]) 
    axs[i].set_title(feature) 

plt.tight_layout() 
plt.show()
```

```python
#calories graph
sns.set_palette("cool")
sns.histplot(data=df['Calories'], x=df['Calories'], kde=True) 
```

```python
#calories&total steps relation
plt.scatter(df['Calories'],df['TotalSteps'], c = '#88c999')
plt.xlabel("Total steps")
plt.ylabel("Calories burnt")
plt.title("calories & total steps relation")
plt.show()

```

```python
#distance&total steps relation
plt.scatter(df['TotalDistance'],df['TotalSteps'], c = 'hotpink')
plt.xlabel("Total distance")
plt.ylabel("Total steps")
plt.title("distance & total steps relation")
plt.show()

```
